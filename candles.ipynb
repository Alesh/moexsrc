```python
import os
import pandas as pd
from moexsrc import session
from moexsrc.dataframes import Market, Ticker, Period

session.TOKEN = os.getenv("APIKEY")
```

```python
moex = Ticker("MOEX")
await moex.candles("5min", begin="2026-01-01", end="2026-02-28")
```

```python
fo = Market("FO")
silv = await fo.get_tickers(assetcode="SILV")
df = pd.concat([await s.candles(Period.ONE_HOUR, begin="2026-01-01", end="2026-02-28") for s in silv])
df.sort_values(["begin", "secid"], ascending=[True, False]).reset_index()
```

```python
df = pd.concat([await s.candles(Period.TEN_MINUTES, latest=2) for s in silv])
df.sort_values(["begin", "secid"], ascending=[True, False]).reset_index()
```
